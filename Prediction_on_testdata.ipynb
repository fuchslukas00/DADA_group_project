```r
library(gbm)

load("../../group_project2/Data_Project2/Project2_data.Rdata")

gbm_fraw <- readRDS("gbm_fraw_model.rds")

# Recreate the same selected features
cor_fraw <- apply(LENS_tas_ann_train$X, 2, function(x)
  cor(x, LENS_tas_ann_train$Y$fraw)
)

top_features <- order(abs(cor_fraw), decreasing = TRUE)[1:500]

# Build test data
test_data <- as.data.frame(LENS_tas_ann_test$X[, top_features])

# Important: match names expected by the GBM model
names(test_data) <- gbm_fraw$var.names

# Predict unseen test set
pred_test <- predict(
  gbm_fraw,
  test_data,
  n.trees = 2000
)

head(pred_test)

write.csv(
  data.frame(fraw_pred = pred_test),
  "gbm_fraw_test_predictions.csv",
  row.names = FALSE
)
```
